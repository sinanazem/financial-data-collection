# Topic

```python
driver = webdriver.Chrome()
driver.get('https://www.goldmansachs.com/insights/topics/index.html')
link_list_topic = [i.get_attribute('href') for i in driver.find_elements(By.CSS_SELECTOR,'.title-link-hover')]
```

```python
def goldmansachs_insight(link):
    
        response = requests.get(link)
        soup = BeautifulSoup(response.text)
        
        # Title
        if soup.find(class_='article-content-page__title article-margin'):
            title = soup.find(class_='article-content-page__title article-margin').text
        else:
            title = 'N/A'
        
        # Date
        if soup.find(class_='article-content-page__date'):
            date = soup.find(class_='article-content-page__date').text
        else:
            date = 'N/A'
        
        # Tag
        if soup.find(class_='article-content-page__topics'):
            raw_tag = soup.find(class_='article-content-page__topics').text.strip()
            tag = re.sub('\s',' ', raw_tag).replace('TOPICS:','').replace('|',',').strip()
        else: 
            tag = 'N/A'
        
        # Content
        if soup.find(class_='article-content-page__content'):
            raw_content = soup.find(class_='article-content-page__content').text
            content = re.sub('\s',' ',raw_content).strip()
        else:
            content = 'N/A'

        return {         
            'index': '',
            'company': 'Goldman Sachs',
            'topic': 'Insight',
            'tag': tag,
            'title': title,
            'date': date,
            'link': link,
            'content': content,
            }
```

```python
goldmansachs_insight_list = []
index = 1

for link in link_list_article:
    goldmansachs_insight_article_dict = goldmansachs_insight(link)
    goldmansachs_insight_article_dict['index'] = index
    goldmansachs_insight_list.append(goldmansachs_insight_article_dict)
    
    index +=1
```
