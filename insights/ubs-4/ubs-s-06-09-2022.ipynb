```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from time import sleep
```

```python
def section_insights_links(main_link):


    service = Service(executable_path=ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service)
    driver.implicitly_wait(5)
    driver.get(main_link)
    
    c = driver.find_element(By.CSS_SELECTOR, "#doc > div:nth-child(12) > div > div.privacysettings__bannerButton > button").click()
    if bool(c) is True:
        c[0]
    driver.find_element(By.CSS_SELECTOR, "#mainnavigation > div.grid-nav.amheader__menu > div.ammainnav__overlayWrapper.verticalshrink__base.ammainnav__overlayWrapper--fullScreen.verticalshrink__base--transition.verticalshrink__base--expanded > div > div > div.contextdisclaimer__base--reimagined > footer > div > div > div > button").click()

    while True:

        try:
            button = driver.find_element(By.CSS_SELECTOR,"#main > div > div:nth-child(2) > div.grid-container.gridcontrol2__gridContainer.gridcontrol2__gridContainer--vertical-spacing-standard > div > div > div > div > div.js-showmore-wrapper > section > div > button")
            driver.execute_script("arguments[0].click();", button)

        except: 
            break

    sleep(7)

    link_list = [i.get_attribute('href') for i in driver.find_elements(By.XPATH,"//*[@class='gridcontrol2__base xxSmall' ][1]//a[@class='sdactivitystreaminsights__linkHl sdactivitystream__itemMainLink']")]
#     title_list = [i.text for i in driver.find_elements(By.XPATH,"//*[@class='gridcontrol2__base xxSmall' ][1]//h3[@class='sdactivitystreaminsights__hl']")]
#     tag_list = [i.text for i in driver.find_elements(By.XPATH, "//*[@class='gridcontrol2__base xxSmall' ][1]//*[@class='sdactivitystreaminsights__meta']//*[@class='sdactivitystreaminsights__metaItem sdactivitystreaminsights__metaItem--leadTags']")]
#     info_list = [i.text for i in driver.find_elements(By.XPATH, "//*[@class='gridcontrol2__base xxSmall' ][1]//*[@class='sdactivitystreaminsights__meta']")]

    driver.quit()
    
    return link_list
```
