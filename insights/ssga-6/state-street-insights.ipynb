```python
from selenium.webdriver.common.action_chains import ActionChains
from time import sleep
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
import json
from pprint import pprint
from tqdm import tqdm
import pandas as pd
```

```python
driver = webdriver.Chrome()
url = "https://www.ssga.com/us/en/intermediary/ic/insights?"
driver.get(url)
sleep(10)
elem = driver.find_element(By.CSS_SELECTOR,'#js-ssmp-clrButtonLabel')
elem.click()
sleep(5)
button = driver.find_element(By.CSS_SELECTOR,'.ssmp-load-more')
for i in range(40):
    driver.execute_script("arguments[0].click();", button)

sleep(5)
articles = driver.find_elements(By.CSS_SELECTOR,'.result-item')
# title = [link.text for link in [i.find_element(By.TAG_NAME,'a') for i in articles]]
link = [link.get_attribute('href') for link in [i.find_element(By.TAG_NAME,'a') for i in articles]]

title_list = [i.text for i in driver.find_elements(By.CSS_SELECTOR,'.title')]
date_list = [i.text for i in driver.find_elements(By.CSS_SELECTOR,'.ssmp-d-mobile-none span:nth-child(1)')]
#author_list = [i.text for i in driver.find_elements(By.CSS_SELECTOR,'.author')]
subject_list = [i.text for i in driver.find_elements(By.CSS_SELECTOR,'.eyebrow a')]
description_list = [i.text for i in driver.find_elements(By.CSS_SELECTOR,'.desc')]



input('exit?')
driver.quit()
```

```python
list_of_dict = []
for title, date, subject, description, l in zip(title_list, date_list, subject_list, description_list, link):
    d = {
        'title': title,
        'date': date,
        'subject': subject,
        'description': description,
        'link': l,
    }
    list_of_dict.append(d)
```

```python
ubs_insight_df = pd.DataFrame(list_of_dict)
```
