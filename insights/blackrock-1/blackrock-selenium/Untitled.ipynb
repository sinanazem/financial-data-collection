 ### 1. Getting a list of all articles by `Selenium`

```python
driver = webdriver.Chrome()
driver.get('https://www.blackrock.com/us/individual/insights')
sleep(10)
elem = driver.find_element(By.ID,'onetrust-reject-all-handler')
elem.click()
button = driver.find_element(By.XPATH,'//*[@id="c1561158932259"]/div/div[2]/div/div/div[3]/button')
for i in range(8):
    
    driver.execute_script("arguments[0].click();", button)

titles = driver.find_elements(By.XPATH,'//h2')
title_list = [i.text for i in titles]

dates = driver.find_elements(By.XPATH, '//a//div//div//span[(((count(preceding-sibling::*) + 1) = 1) and parent::*)]')
date_list = [i.text for i in dates]

by = driver.find_elements(By.CSS_SELECTOR,'span:nth-child(4)')
by_list = [i.text for i in by] 

descriptions = driver.find_elements(By.XPATH,'//*[contains(concat( " ", @class, " " ), concat( " ", "description", " " ))]')
description_list = [i.text for i in descriptions]

links = driver.find_elements(By.CSS_SELECTOR,'.skip-animation')
link_list = [i.get_attribute('href') for i in links]

list_of_dict = []
for title,date,b,description,link in zip(title_list,date_list,by_list,description_list,link_list):
    d = {
        'company': 'BlackRock', 
        'topic': 'Insight',
        'title':title,
        'link':link,
        'description':description,
        'date':date,
        'by':b
    }
    list_of_dict.append(d)




# button = driver.find_element(By.XPATH,'//*[@id="c1561158932259"]/div/div[2]/div/div/div[3]/button')
# driver.execute_script("arguments[0].click();", button)

input('exit?')
driver.quit()
```

### 3. Save in `json` file

```python
with open('blackrock_insight_1.json', 'a') as f:
    json.dump(list_of_dict, f, indent=4)
```

### 4. Create `pandas` dataframe

```python
blackrock_insight_1 = pd.DataFrame(list_of_dict)
```

```python
blackrock_insight = blackrock_insight_1.merge(blackrock_insight_2, left_on='link',right_on='url',how='left').drop(columns=['url'])
```
