```python
# ticker => is a abbreviation of the name of the companies which we feed to the function
def get_earning_urls(ticker):
    url = f"https://www.fool.com/quote/nasdaq/{ticker}/"
    base_url='https://www.fool.com'
    print(url)
    earning_call_links   = []
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, "html.parser")
        html_link_list = soup.find(id="earnings-transcript-container").find_all("a")

        
        for link in html_link_list:
            earning_call_links.append(base_url + (link['href']))
    except:
        pass
    return earning_call_links
    

import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

table = soup.find('table', {'class': 'wikitable sortable'})
tickers = []

for row in table.find_all('tr')[1:]:
    ticker = row.find_all('td')[0].text.strip()
    name = row.find_all('td')[1].text.strip()
    tickers.append((ticker, name))
    
list_of_link = [get_earning_urls(i) for i in [i[0] for i in tickers]]
```

```python
def content_producer(url):
    response_url = requests.get(url).text
    soup_myurl = BeautifulSoup(response_url, "html.parser")
    html_content = soup_myurl.find(class_ = "tailwind-article-body").text
    html_title = soup_myurl.find(class_ = "font-medium text-gray-1100 leading-42 md:text-h1").text
    html_date = soup_myurl.find(id="date").text
    html_current_price = soup_myurl.find(class_ = "text-sm font-medium lg:text-right lg:pl-8px text-gray-1100 p-8px lg:text-lg").text
    crawl_time = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    
    
    url_dictionary = {
        'title': html_title,
        'date' : html_date,
        'crawl_time': crawl_time,
        'current_price': html_current_price,
        'content': html_content,
        
    }
    
    return url_dictionary
```
